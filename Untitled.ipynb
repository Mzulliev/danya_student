## Знакомство с Jupyter Notebook

Для запуска jupyter notebook вводим в терминал команду:
```bash
jupyter-notebook
```

- Выполнить ячейку и перейти далее ```Shift + Enter```
- Выолнить ячейку без перехода ```Ctr + Enter```
- Добавить ячейку выше ```A```
- Добавить ячейку ниже ```B```
- Вырезать ячейку ```X```
- Копировать ячейку ```C```
- Вставить ячейку ```V```
- Удалить ячейку ```DD```
- Изменить на Markdown ```M```
- Изменить на код ```Y```
- Отменить удаление ```Z```

      Примечание: Переключение в командный режим - ```Esc```, в режим редактирования - ```Enter```

### Markdown в Jupyter
- ```#``` Заголовок первого уровня (всего их 6)
- ```**жирный**``` или ```__жирный__``` или ```<b>жирный</b>```
- ```*курсив*``` или ```_жирный_``` или ```<i>жирный</i>```
- Для формул: ```$ формула $``` (inline) или ```$$ формула $$``` (block)
- Списки: ```- элемент```, ```1. элемент```


## Git и GitHub
1. Открываем терминал
2. Переходим в папку проекта

   ```сd путь/до/вашего/проекта```
   
3. Инициализируйте git-репозиторий, если еще не сделали:
    
    ```git init```

4. Добавление всех файлов в индекс:

    ```git add .```

5. Сделайте новый коммит:

    ```git commit -m "Initital commit"```

6. Создайте репозиторий на GitHub (через сайт - кнопка "New repository"). После создания скопируйте ссылку на репозиторий.
7. Свяжите локальный репозиторий с GitHub

    ```git remote add origin https://github.com/your_username/your_repo.git```

8. Отправьте ваш проект на GitHub

    ```git push -u orogin main```

После этого проект будет загружен на GitHub и синхронизирован через терминал.

**Основные команды Git для терминала**

- *git status* - текущее состояние
- *git add .* - добавить все файлы
- *git commit* -m "сообщение" - коммит с комментарием
- *git push* - отправить изменения на GitHub
- *git pull* - забрать изменения с GitHub
